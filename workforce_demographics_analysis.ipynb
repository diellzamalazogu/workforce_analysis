For this question I picked Sex as a categorical data, which I'm going to explore how this variable changes with Age

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

#Transforming Sex to a Binary variable in order to perform logistic regression
mydata['SexClass'] = np.where(mydata['Sex']==" Male", 1, 0)

d=mydata[['Age']]
data=d.to_numpy()
t=mydata['SexClass']
target=t.to_numpy()


X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.6)
#train logistic regression model on training data
logreg = LogisticRegression()
a=logreg.fit(X_train, y_train)
y_pred=logreg.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
print("Model score is "+str(logreg.score(X_train, y_train)))
```

Since we have a categorical data "Sex", we use logistical regression to conclude the relationship between the two variables; categorical variable "Sex" and numerical variable "Age".From the classification report and model score we can conclude that variable that I selected (Sex) does not appear to depend strongly on age.

```python
from sklearn import model_selection

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

d=mydata[['Age']]
data=d.to_numpy()
t=mydata['Hours per week']
target=t.to_numpy()

# split data set
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.6)

# train regression model on training data
regr = LinearRegression(fit_intercept=True)
a=regr.fit(x_train,y_train)
y_pred= regr.predict(x_test)

# regression model on test data
b=regr.fit(x_test,y_test)
y_pred2= regr.predict(x_train)
a=r2_score(y_train,y_pred2)
b=r2_score(y_test, y_pred)
print("Mean squared error for validation data is: %.2f"% mean_squared_error(y_test, y_pred))
print("Mean squared error for training data is: %.2f"% mean_squared_error(y_train, y_pred2))
print("The model score for training data is: "+str(a))
print("The model score for validation data is: "+str(b))
residual = y_test - y_pred
fig, ax = plt.subplots(figsize=(6,2.5))
ax.scatter(residual, y_pred)
ax.set_title("Plot of residuals against predicted values for training data")
ax.set_ylabel("residuals")
ax.set_xlabel("predicted values")
```

The model score for both training and validation data is very low, meaning the proportion of the variance of the dependent variable that is explained by the independent variables in our case is very low, more precisly around 0.4% for both cases. Also, we can also see that the Mean Squared error is high, meaning that our data points are not that close to the fitted line. Furthermore, the residual plot patterns suggest that there is no linear relationship.
